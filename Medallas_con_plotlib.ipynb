```python
import matplotlib.pyplot as plt
import pandas as pd

atletas1df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/athlete_events-1.csv")

atletas1df = atletas1df[atletas1df.Medal.notnull()]
atletas1df.head()

medallas_por_pais = atletas1df["Team"].value_counts()

#Para mostrar países usamos .head() sobre la tabla, como siempre. Pero para mostrar los 10 primeros usamos .head(10)

medallas_por_pais = medallas_por_pais.head(10)

plt.suptitle("Países con más medallas de toda la historia")
plt.xlabel("Países")
plt.ylabel("Cantidad de medallas")
medallas_por_pais.plot(kind="bar",color="darkgreen")
plt.show()
print("")
print("")
print("")
ultimo_año = atletas1df["Year"].max()

#Acá estamos filtrando por los países que tienen medallas en el último año.

latest = atletas1df[atletas1df["Year"]==ultimo_año]

#Acá armamos el ranking de nuevo, pero con los países mas ganadores en el último año. Por eso por ejemplo la Unión Soviética no figura, es re contra ganadora pero no en el último año.
medallas_ult_año = latest["Team"].value_counts()

medallas_ult_año.head(10).plot(kind="pie")

#Para indicarle cual columna de la base va a servir de etiqueta hacemos el paso siguiente:
etiquetas = ["Teams"]
plt.suptitle("Países con más medallas en el último año")
plt.legend("Teams")
plt.legend(loc="upper right")
plt.legend(fontsize=9)
leyenda = plt.legend()
leyenda.set_alpha(0.2)
plt.show()

```
