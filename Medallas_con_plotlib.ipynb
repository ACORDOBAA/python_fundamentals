```python
import matplotlib.pyplot as plt
import pandas as pd

atletas1df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/athlete_events-1.csv")

atletas1df = atletas1df[atletas1df.Medal.notnull()]
atletas1df.head()

medallas_por_pais = atletas1df["Team"].value_counts()

#Para mostrar pases usamos .head() sobre la tabla, como siempre. Pero para mostrar los 10 primeros usamos .head(10)

medallas_por_pais = medallas_por_pais.head(10)

plt.suptitle("Pases con ms medallas de toda la historia")
plt.xlabel("Pases")
plt.ylabel("Cantidad de medallas")
medallas_por_pais.plot(kind="bar",color="darkgreen")
plt.show()
print("")
print("")
print("")
ultimo_ao = atletas1df["Year"].max()

#Ac estamos filtrando por los pases que tienen medallas en el ltimo ao.

latest = atletas1df[atletas1df["Year"]==ultimo_ao]

#Ac armamos el ranking de nuevo, pero con los pases mas ganadores en el ltimo ao. Por eso por ejemplo la Unin Sovitica no figura, es re contra ganadora pero no en el ltimo ao.
medallas_ult_ao = latest["Team"].value_counts()

medallas_ult_ao.head(10).plot(kind="pie")

#Para indicarle cual columna de la base va a servir de etiqueta hacemos el paso siguiente:
etiquetas = ["Teams"]
plt.suptitle("Pases con ms medallas en el ltimo ao")
plt.legend("Teams")
plt.legend(loc="upper right")
plt.legend(fontsize=9)
leyenda = plt.legend()
leyenda.set_alpha(0.2)
plt.show()

```
