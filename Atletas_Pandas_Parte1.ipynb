```python
import pandas as pd
import numpy as np

#La "r" antes de las comillas del comando de lectura del archivo .csv está convirtiendo toda la ruta en un formato RAW que permite que sea leído. Sino habría que poner doble barra en cada barra para que tome OK el código.

atletasdf = pd.read_csv(r"athlete_events.csv")
#print(atletasdf.head())

#-------------------------------------- Punto 3
#Hay valores que vienen NaN en Medallas (y en otros campos también), a esos los tenemos que limpiar

atletas_not_nan = atletasdf[atletasdf.Medal.notnull()]
print("")
print("Tabla limpia de atletas sin medallas")
print("")
print(atletas_not_nan.head())
#--------------------------------------- Punto 4
#Este paso me busca el más veterano que no tiene valor NaN en la columna Medallas.
print("")
index_edad = atletas_not_nan["Age"].idxmax()

#Con la funcion .loc puedo buscar en base al nombre de la columna, usando .iloc necesito el número de la columna o datos tipo int.
print("Punto 4: Atleta condecorado más veterano")
print("")
print(atletas_not_nan.loc[index_edad])

#--------------------------------------------------------Punto 5: Ahora armamos otro índice para encontrar al más joven dentro de la nueva base de atletas oro, y después lo imprimimos

#Acá filtramos las filas que nos devuelven "True" en la columna Medal, cuando el valor es Gold
filtro_oro = atletas_not_nan["Medal"] == "Gold"
filtro_oro.head()

#Y acá filtramos la base atletas not nan, a través del filtro de medallas = Oro. Y pedimos que nos imprima la tabla
#atletas_oro = atletas_not_nan[filtro_oro]
#atletas_oro.head()


index_edad2 = atletas_oro["Age"].idxmin()
print("")
print("Punto 5: Atleta más joven condecorado con oro")
print("")
print(atletas_oro.loc[index_edad2])

#----------------------------------------------------------- Punto 7

mas_medallas = atletas_not_nan["Name"].value_counts().idxmax()
print("")
print("Punto 7: Atleta más condecorado")
print("")
print(mas_medallas)

#----------------------------------------------------------- Punto 8: Creación de .csv con la data del atleta más condecorado (Phelps)

index_phelps = atletas_not_nan["Name"] == "Michael Fred Phelps, II"
atletas_phelps = atletas_not_nan[index_phelps]
atletas_phelps.to_csv("phelps.csv")





```
