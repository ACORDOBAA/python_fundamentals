```python
import pandas as pd
import matplotlib.pyplot as plt

baseamazon = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Amazon Customer Behavior Survey.csv")
baseamazon.head

#Edad que más se repite?
edad1 = baseamazon["age"].value_counts().idxmax()
print("La edad que más se repite es:", edad1, "años")

#Género que más se repite
genero1 = baseamazon["Gender"].value_counts().idxmax()
print("El género que más se repite es:", genero1)
print("")

#Cuáles son los motivos de abandono?
print("")
print("\033[4mCAUSAS DE ABANDONO DE CARRITOS\033[0m")
print("")
print("Y todos los motivos de abandono de carritos son:")
motivos = baseamazon.Cart_Abandonment_Factors.value_counts()
print("")
print("")
plt.suptitle("Motivos de abandono de carrito")
plt.xlabel("Motivos")
motivos.plot(kind="bar",color="darkgreen")
plt.show()
print("")
print("")
print("")

#Cuál es el motivo de abandono top?
print("El motivo principal de abandono es:", baseamazon.Cart_Abandonment_Factors.describe().top)
print("")
print("")


#Cuáles son las categorías de productos principales?
print("\033[4mPRODUCTOS PRINCIPALES\033[0m")
print("")
print("Las 5 categorías de productos principales son:")
print("")
principales = baseamazon["Purchase_Categories"].value_counts()
principales1 = principales.head(5)

principales.head(5).plot(kind="pie")
#etiquetas = ["Purchase_Categories"]
plt.suptitle("Categorías mas buscadas")
#plt.legend("Purchase_Categories")
#plt.legend(loc="upper right")
#plt.legend(fontsize=9)
#leyenda = plt.legend()
#leyenda.set_alpha(0.2)
plt.show()
print("")
print("")

#Principales factores que influencian la frecuencia de compra

influencias = baseamazon["Service_Appreciation"].value_counts()
influencias = principales.head(5)
print("\033[4mINFLUENCIA DE COMPRA\033[0m")
print("")
print("Los principales factores que influencian una mayor frecuencia de compra son:   (cantidad de registros)")
print(influencias)
print("")
print("")

#Satisfacción del cliente
print("\033[4mSATISFACCIÓN DEL CLIENTE\033[0m")
print("")
print("Del 1 al 5, la satisfacción del cliente se compone de la siguiente manera:")
satisfaccion = baseamazon.Shopping_Satisfaction.value_counts()
print("")
print("")
plt.suptitle("")
plt.xlabel("Grado de satisfacción")
satisfaccion.plot(kind="bar",color="skyblue")
plt.show()
print("")
print("")
print("")

#Principales áreas de oportunidad
print("\033[4mÁREAS DE OPORTUNIDAD\033[0m")

areas_oportunidad = baseamazon.Improvement_Areas.value_counts()
areas_1 = baseamazon["Improvement_Areas"].value_counts().idxmax()
areas_oportunidad.head(5).plot(kind="pie")
plt.suptitle("Áreas de oportunidad")

plt.show()
print("")
print("")

print("Y el área de oportunidad número 1 es:", "\033[1m" + areas_1 + "\033[0m" )




```
