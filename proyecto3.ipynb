```python
import pandas as pd
import numpy as np
import sys

basemro1 = "/content/base mro feb19 copia.xlsx"
basemro2 = "/content/base mro mar19 copia.xlsx"

df1 = pd.read_excel(basemro1)
df2 = pd.read_excel(basemro2)

def calcular_diferencia(row):
    if row["Provision Nuevo Criterio_df2"] == 0:
        return row["Provision Nuevo Criterio_df1"] * -1
    elif row["Provision Nuevo Criterio_df1"] == 0:
        return row["Provision Nuevo Criterio_df2"]
    else:
        return abs(row["Provision Nuevo Criterio_df2"] - row["Provision Nuevo Criterio_df1"])

df_resultado = pd.merge(df1, df2, on="Concatenado", suffixes=("_df1", "_df2"))

df_resultado["Variac"] = df_resultado.apply(calcular_diferencia, axis=1)

base_mro_cruce = df_resultado

filtro_variac_cruce = (base_mro_cruce["Variac"] >= 1) | (base_mro_cruce["Variac"] <= -1)
base_mro_cruce_filtrada = base_mro_cruce[filtro_variac_cruce]

base_mro_cruce_filtrada = base_mro_cruce_filtrada.sort_values(by='Variac', ascending=False)

#COMIENZO DE COMENTARIOS PARA .TXT

#Recuento de Materiales para comentario
recuento_mros = len(base_mro_cruce_filtrada)

with open('comentarios variaciones.txt', 'w') as f:

    original_stdout = sys.stdout
    sys.stdout = f

    print("La cantidad de MROs que sufrieron modificaciones en su stock provisionado fueron:", recuento_mros, "materiales, según la siguiente apertura por almacenes:")
    print("")

    #Top 20 de Materiales que aportan a la variación de la provisión
    columnas_deseadas = ["Material_df2", "Descripcion Gr. Art._df2","Cant_df2","UMB_df2","Precio MG_df2","Variac" ]

    print("Los 20 Materiales que más aportaron a la variación de la provisión fueron:")
    print(base_mro_cruce_filtrada.head(20)[columnas_deseadas])
    print("")

    #Apertura de Materiales por Almacenes
    print("Apertura de MROs por Almacenes")
    num_mros_SR01 = base_mro_cruce_filtrada[base_mro_cruce_filtrada["Area_df2"] == "SR01"].count()["Area_df2"]
    print("MROs afectados en Almacén Rosario:", num_mros_SR01, "materiales, siendo los principales:")
    print("")

    num_mros_SC01 = base_mro_cruce_filtrada[base_mro_cruce_filtrada["Area_df2"] == "SC01"].count()["Area_df2"]
    print("MROs afectados en Almacén Canning:", num_mros_SC01, "materiales, siendo los principales:")
    print("")

    num_mros_SS01 = base_mro_cruce_filtrada[base_mro_cruce_filtrada["Area_df2"] == "SS01"].count()["Area_df2"]
    print("MROs afectados en Almacén San Nicolás:", num_mros_SS01, "materiales, siendo los principales:")
    print("")

    num_mros_SH01 = base_mro_cruce_filtrada[base_mro_cruce_filtrada["Area_df2"] == "SH01"].count()["Area_df2"]
    print("MROs afectados en Almacén Haedo:", num_mros_SH01, "materiales, siendo los principales:")
    print("")

    num_mros_SE01 = base_mro_cruce_filtrada[base_mro_cruce_filtrada["Area_df2"] == "SE01"].count()["Area_df2"]
    print("MROs afectados en Almacén Ensenada:", num_mros_SE01, "materiales, siendo los principales:")
    print("")

    num_mros_SF01 = base_mro_cruce_filtrada[base_mro_cruce_filtrada["Area_df2"] == "SF01"].count()["Area_df2"]
    print("MROs afectados en Almacén FV:", num_mros_SF01, "materiales, siendo los principales:")
    print("")


    sys.stdout = original_stdout

#SALIMOS DEL MODO COMENTARIOS PARA .TXT

#archivo_excel_filtrado = "/content/base_mro_cruce.xlsx"
#df_resultado.to_excel(archivo_excel_filtrado, index=False)




```
